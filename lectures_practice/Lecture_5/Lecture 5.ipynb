```python
import mpl_toolkits.basemap as bm
import matplotlib.pyplot as plt
import numpy as np


theMap = bm.Basemap(projection='ortho',
                   lat_0=48.2, lon_0=16.4,
                   llcrnry=-4000000.0, llcrnrx=-4000000.0,
                   urcrnry=4000000.0, urcrnrx=4000000.0,
                   resolution='i', area_thresh=1000)


theMap.drawcountries()
theMap.drawcoastlines()

theMap.fillcontinents(color='coral', lake_color='aqua')
theMap.drawmapboundary(fill_color='aqua')

theMap.drawparallels( np.arange(0, 90, 15))
theMap.drawmeridians( np.arange(-120, 120, 30))

plt.show()

# start watching at 28:26 Lecture 5
```

```python
# Single map, two subplots
theMap = bm.Basemap(projection='ortho',
                   lat_0=48.2, lon_0=16.4,
                   llcrnry=-1000000.0, llcrnrx=-1000000.0,
                   urcrnry=1000000.0, urcrnrx=1000000.0,
                   resolution='i', area_thresh=1000)

fig = plt.figure()

# We are changing state to subplot 1
ax1 = fig.add_subplot(121)

theMap.drawcountries()
theMap.drawcoastlines()
theMap.fillcontinents(color='coral', lake_color='aqua')
theMap.drawmapboundary(fill_color='aqua')
# theMap.bluemarble()

# We are changing state to subplot 1
ax1 = fig.add_subplot(122)

theMap.drawcountries()
theMap.drawcoastlines()

# Ooops, let's go back to subplot 1 and draw a grid
ax1 = fig.add_subplot(121)
theMap.drawparallels( np.arange(0, 360, 10))
theMap.drawmeridians( np.arange(-90, 90, 10))

plt.show()
```
